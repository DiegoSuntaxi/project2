```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import os
import pandas as pd 
from io import StringIO
import numpy as np
from bs4 import BeautifulSoup
```

```python
# Path to chromedriver executable
chromedriver_path = "chromedriver.exe"

# Initialize ChromeDriver with Selenium
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)
```

```python
URL = "http://localhost:8080/"
```

```python
driver.get(URL)
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_html(selector, timeout=10):
    wait = WebDriverWait(driver, timeout)
    
    elements = wait.until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, selector))
    )
    
    return [el.get_attribute("outerHTML") for el in elements]
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

items = soup.select("div")

for item in items:
    print(item.text)
```

```python
h3_tags = soup.find_all("h3")
print("H3 tags:")
for tag in h3_tags:
    print(tag.text)
```

```python
p_horas = soup.find_all("p", class_="card-text", id="today-hours")
print("P horas:")
for tag in p_horas:
    print(tag.text)
```

```python
horas_limpias = []

for tag in p_horas:
    texto = tag.text.strip()
    limpio = texto.replace("Hoy:", "").replace("🕒", "").strip()
    horas_limpias.append(limpio)

for h in horas_limpias:
    print(h)
```
