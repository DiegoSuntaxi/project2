```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import os
import pandas as pd 
from io import StringIO
import numpy as np
from bs4 import BeautifulSoup
```

```python
# Path to chromedriver executable
chromedriver_path = "chromedriver.exe"

# Initialize ChromeDriver with Selenium
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)
```

```python
URL = "http://localhost:8080/"
```

```python
driver.get(URL)
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_html(selector, timeout=10):
    wait = WebDriverWait(driver, timeout)
    
    elements = wait.until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, selector))
    )
    
    return [el.get_attribute("outerHTML") for el in elements]
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

items = soup.select("div")

for item in items:
    print(item.text)
```

```python
h3_tags = soup.find_all("h3")
print("H3 tags:")
for tag in h3_tags:
    print(tag.text)
```

```python
horas_limpia = [
    tag.text.replace("Hoy:", "").replace("🕒", "").strip()
    for tag in soup.find_all("p", class_="card-text", id="today-hours")
]

for h in horas_limpia:
    print(h)
```

```python
direcciones_limpias = [
    tag.text.replace("Dirección:", "").strip()
    for tag in soup.find_all("p", class_="card-text", id="address")
]

for d in direcciones_limpias:
    print(d)
```

```python
calificaciones_limpias = [
    tag.text.replace("Calificación:", "").replace("⭐", "").strip()
    for tag in soup.find_all("p", class_="card-text", id="rating")
]

for c in calificaciones_limpias:
    print(c)
```

```python
usuarios_limpios = [
    tag.text.replace("Usuarios:", "").replace("👥", "").strip()
    for tag in soup.find_all("p", class_="card-text", id="users")
]

for u in usuarios_limpios:
    print(u)
```

```python
def limpiar(texto, *remplazos):
    for r in remplazos:
        texto = texto.replace(r, "")
    return texto.strip()

direcciones = [
    limpiar(tag.text, "Dirección:")
    for tag in soup.find_all("p", class_="card-text", id="address")
]

horarios = [
    limpiar(tag.text, "Hoy:", "🕒")
    for tag in soup.find_all("p", class_="card-text", id="today-hours")
]

ratings = [
    limpiar(tag.text, "Calificación:", "⭐")
    for tag in soup.find_all("p", class_="card-text", id="rating")
]

usuarios = [
    limpiar(tag.text, "Usuarios:", "👥")
    for tag in soup.find_all("p", class_="card-text", id="users")
]
```

```python
import pandas as pd

df = pd.DataFrame({
    "direccion": direcciones,
    "horario": horarios,
    "rating": ratings,
    "usuarios": usuarios
})

print(df)
```

```python
df.to_csv("lugares.csv", index=False, encoding="utf-8-sig")
```
